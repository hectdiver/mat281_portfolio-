El **cáncer de mama** es una proliferación maligna de células epiteliales en los conductos o lobulillos mamarios. Esta enfermedad clonal surge cuando una célula adquiere, a través de mutaciones, la capacidad de dividirse sin control, formando un tumor que puede invadir tejidos vecinos y propagarse a otras partes del cuerpo.

El conjunto de datos `BC.csv` contiene información sobre pacientes con tumores (benignos o malignos) y diversas características del tumor, calculadas a partir de imágenes digitalizadas de aspirados con aguja fina (FNA) de masas mamarias. Estas características describen los núcleos celulares en la imagen y permiten diferenciar entre tumores benignos y malignos.

A continuación, cargamos el conjunto de datos:

```python
# Importar librerías
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Importar herramientas de Scikit-learn
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Configuración de gráficos
%matplotlib inline
sns.set_palette("deep", desat=0.6)
sns.set(rc={'figure.figsize': (11.7, 8.27)})

# Cargar y preparar los datos
df = pd.read_csv("https://raw.githubusercontent.com/fralfaro/MAT281_2024/main/docs/labs/data/BC.csv")
df.set_index('id', inplace=True)

# Transformación de la variable objetivo
df['diagnosis'] = df['diagnosis'].replace({'M': 1, 'B': 0}).astype(int)

# Visualizar las primeras filas del DataFrame
df.head()
```

2. Normalizar las variables numéricas con el método **StandardScaler**.

```python
# FIXME
# Variables:
X = df.drop('diagnosis', axis=1)
y = df['diagnosis']

# Normalizar con StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# transformar a dataframe
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns, index=X.index)

# Visualizar
X_scaled_df.head()
```

3. Realizar un método de reducción de dimensionalidad visto en clases.

```python
# FIXME
#metodo de reduccion de dimensionalidad con PCA
pca = PCA(n_components=2) #reducir a 2 componentes
X_pca = pca.fit_transform(X_scaled) #Aplicar PCA a conhjunto de datos escalado obtenidos en punto 2
X_pca_df = pd.DataFrame(X_pca, columns=[f'PC{i+1}' for i in range(2)], index=X.index) #transformar a dataframe
```

4. Aplique al menos tres modelos de clasificación distintos. Para cada uno de los modelos escogidos, realice una optimización de los hiperparámetros. además, calcule las respectivas métricas. Concluya.

Se aplicaran los metodos de regresion logistica, arboles de decision y random forest. Para cada modelos se buscaran hiperparámetros con el metodo de GrindSearchCV.

```python
# Regresion Logistica:


# datos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_pca_df, y, test_size=0.3, random_state=42, stratify=y)

# iteracion modelo de Logistic Regression
logreg = LogisticRegression(max_iter=1000)

# hiperparámetros Optimizacion
param_grid = {'C': [0.1, 1, 10, 100],  'solver': ['liblinear', 'lbfgs']}

# Buscar hiperparámetros con GridSearchCV
grid_search = GridSearchCV(logreg, param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)
y_mejor = grid_search.best_estimator_
y_pred = y_mejor.predict(X_test)

# Metricas de evaluación
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

# Imprimir metricas de evaluación
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)






```

```python
#ARBOLES DE DECISION

from sklearn.tree import DecisionTreeClassifier

# datos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_pca_df, y, test_size=0.3, random_state=42, stratify=y)
dtree = DecisionTreeClassifier(random_state=42)

# hiperparámetros Optimizacion
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 5, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 5]
}

# Buscar hiperparámetros con GridSearchCV
grid_search = GridSearchCV(dtree, param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)
y_mejor = grid_search.best_estimator_
y_pred = y_mejor.predict(X_test)

# Metricas de evaluación
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

# Imprimir Metricas de evaluación
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```

```python
#Random Forest:

# atos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_pca_df, y, test_size=0.3, random_state=42, stratify=y)

rf = RandomForestClassifier(random_state=42)

# hiperparámetros OPTIMIZACION
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [5, 10, 20],
    'criterion': ['gini', 'entropy']
}

# Realizar la búsqueda de hiperparámetros con GridSearchCV
grid_search = GridSearchCV(rf, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(X_train, y_train)
y_mejor = grid_search.best_estimator_
y_pred = y_mejor.predict(X_test)

# Metricas de evaluación
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

# Imprimir metricas de evaluación

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```

Se observa que todas las metricas de evaluacion tienen valores cercanos a 1 validando la aprobacion de los modelos sin embargo el método de Random Forest tiene el Accuracy , Recall y F1 score más alto que los otros modelos por lo indica que es el modelo con mejor balance entre precision y detectar casos positivos. En conclusión se puede decir que de los 3 metodos de clasificacion aplicados el mejor es Random Forest.
