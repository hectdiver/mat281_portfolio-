El conjunto de datos, denominado `company_sales_data.csv`, recoge las ventas mensuales de distintos productos de una empresa, ofreciendo una visión general de las unidades vendidas y los beneficios generados. Entre los productos incluidos se encuentran cremas faciales, jabones de baño y artículos para el cuidado capilar. Este conjunto de datos es útil para analizar el rendimiento de ventas y el comportamiento del mercado a lo largo del tiempo.

### Descripción de los Datos:

1. **month_number**: Número correspondiente al mes del año (1 para enero, 2 para febrero, etc.).
2. **facecream**: Cantidad de unidades de crema facial vendidas durante el mes.
3. **facewash**: Cantidad de unidades de limpiador facial vendidas durante el mes.
4. **toothpaste**: Cantidad de unidades de pasta de dientes vendidas durante el mes.
5. **bathingsoap**: Cantidad de unidades de jabón de baño vendidas durante el mes.
6. **shampoo**: Cantidad de unidades de champú vendidas durante el mes.
7. **moisturizer**: Cantidad de unidades de crema hidratante vendidas durante el mes.
8. **total_units**: Total de unidades vendidas de todos los productos en un mes dado.
9. **total_profit**: Ganancia total obtenida por la venta de todos los productos durante el mes, expresada en la moneda correspondiente.



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(rc={'figure.figsize':(10,8)})
```

```python
# cargar datos
df = pd.read_csv("https://raw.githubusercontent.com/fralfaro/MAT281_2024/main/docs/labs/data/company_sales_data.csv")
df.head()
```

El **objetivo principal** es extraer la mayor cantidad de información posible de este conjunto de datos. Para lograrlo, se deben abordar las siguientes tareas. Para cada pregunta, deberá obtener una conclusión basada en los gráficos generados. Al finalizar todas las preguntas, se espera una conclusión general del análisis.

> **Nota:** Se permite el uso de las librerías Matplotlib o Seaborn para la visualización y análisis de los datos.




1. Lea la columna **`total_profit`** correspondiente a todos los meses y represente los datos utilizando un gráfico lineal y un gráfico de dispersión.


```python
#### GRAFICO LINEAL#####
x = df['month_number']
y = df['total_profit']

#plotear
plt.plot(x,y)
plt.xlabel('Mes')
plt.ylabel('Beneficio Total')
plt.title('Beneficio Total de cada Mes')
plt.show
```

```python
#### GRAFICO DE DISPERSION #####
x = df['month_number']
y = df['total_profit']

plt.scatter(x,y)
plt.xlabel('Mes')
plt.ylabel('Beneficio Total')
plt.title('Beneficio Total de cada Mes')
plt.show
```

Conclusión local: se observa graficamente la ganancia total para cada uno de los meses, el mes 2 es el que tiene un menor beneficio total, mientras que el mes 11 es que tiene mayor cantidad de beneficio total.


2. Lea los datos de ventas de todos los productos y represéntelos en un gráfico de líneas múltiples.


```python
#### GRAFICO DE lineas mixtas para venta de los productos#####

plt.figure(figsize=(10,6))

plt.plot(df['month_number'], df['facecream'], label='facecream')
plt.plot(df['month_number'], df['facewash'], label='facewash')
plt.plot(df['month_number'], df['toothpaste'], label='toothpaste')
plt.plot(df['month_number'], df['bathingsoap'], label='bathingsoap')
plt.plot(df['month_number'], df['shampoo'], label='shampoo')
plt.plot(df['month_number'], df['moisturizer'], label='moisturizer')

plt.xlabel('Mes')
plt.ylabel('Ventas')
plt.title('Ventas de Productos Mensuales')
plt.legend()
plt.show()
```

Conclusión local: se observa que el producto con mayores ventas corresponde a bathingsoap , en donde incluso se observa que a fin de año las ventas crecen. El producto con menos ventas corresponde a moisturizer con ventas bajo 2000 durante todo el año
.

3. Lea los datos de ventas de **`facecream`** y **`facewash`** y represéntelos utilizando un gráfico de barras.

```python
plt.figure(figsize=(10,6))
plt.bar(df['month_number'], df['facecream'], label='facecream')
plt.bar(df['month_number'], df['facewash'], label='facewash')
plt.xlabel('Mes')
plt.ylabel('Ventas')
plt.title('Ventas de Facecream y Facewash Mensuales')
plt.legend()
plt.show()

totalfacecream = sum(df['facecream'])
totalfacewash = sum(df['facewash'])

#graficar valores totales
plt.figure(figsize=(10,6))
plt.bar(1, totalfacecream, label='facecream')
plt.bar(2, totalfacewash, label='facewash')
plt.xlabel('Mes')
plt.ylabel('Ventas')
plt.title('Ventas de Facecream y Facewash Mensuales')
plt.legend()
plt.show()
```

Conclusión Local: en ambos graficos se observa que las ventas de facecream superan significativamente a las ventas de facewash.



4. Lea los datos de ventas de todos los productos y represéntelos utilizando un gráfico de caja (*box-plot*).


```python
#leer datos de ventas de cada producto, y representar con box-plot

products=['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
ventas = df[products]
plt.figure(figsize=(10,6))
sns.boxplot(data=ventas)
plt.xlabel('Productos')
plt.ylabel('Ventas')
plt.title('Ventas de Productos Mensuales')
plt.show()
```

Conclusión Local: el gráfico de caja realizado permite comparar como se distribuyen las ventas de cada producto. se observa que bathingsoap tiene un rango de ventas alto mientas que facewash y moisturizer tienen medianas y rangos bajos por lo que son los con menores ventas.


5. Calcule las ventas totales del año pasado para cada producto y represéntelas utilizando un gráfico circular.

```python
plt.figure(figsize=(10,6))
total_ventas = df[['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']].sum()
plt.pie(total_ventas, labels=total_ventas.index, autopct='%1.1f%%')
plt.title('Ventas Totales por Producto')
plt.show()
```

Conclusion Local: El grafico circular demuestra claramente que el producto bathingsoap es el mas vendi mientras que tanto para facewash como moisturizer corresponden a los productos menos vendidos.


 Conclusion General: Los graficos ayudan bastante a interpretar y analizar los datos de la tabla. En todos los graficos se observa una clara superioridad del producto bathungsoap en la cantidad de ventas, mientras que moisturizer y facewash son los menos vendidos.

 En los gráficos de 1 se puede observar un incremento de los beneficios totales a medida que pasan los meses, para 2. se obvserva incremento de las ventas, mientras que para el resto de graficos se observa la diferencia entre todos los productos.
