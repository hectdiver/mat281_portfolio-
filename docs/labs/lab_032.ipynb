3. Desarrolle una función `resumen_df(df)` para encontrar el total de elementos distintos y vacíos por columnas.

```python
# respuesta
def resumen_df(df):
    """
    Función para generar un resumen de un DataFrame que incluye
    el número de elementos distintos y vacíos por columna.

    Args:
    df (pd.DataFrame): DataFrame a resumir.

    Returns:
    pd.DataFrame: DataFrame resumen con el nombre de las columnas,
                  cantidad de elementos distintos y cantidad de elementos vacíos.
    """
    # Crear un DataFrame de resultado con los nombres de las columnas
    nombres = df.columns
    result = pd.DataFrame({'nombres': nombres})

    # Calcular el número de elementos distintos por columna
    result['elementos_distintos'] = df.nunique().values

    # Calcular el número de elementos vacíos (NaN) por columna
    result['elementos_vacios'] =  df.isna().sum().values

    return result
```

```python
# retornar
resumen_df(df)
```

```python
# respuesta
america = ['ARG', 'ATG', 'BLZ', 'BOL', 'BRA', 'CAN', 'CHL', 'COL', 'CRI',
       'CUB', 'DOM', 'ECU', 'GRD', 'GTM', 'GUY', 'HND', 'HTI', 'JAM',
       'MEX', 'NIC', 'PAN', 'PER', 'PRY', 'SLV', 'SUR', 'TTO', 'URY',
       'USA', 'VEN']

df_america =  pd.DataFrame()


```

```python
#a)Mediante un ciclo for el pais con mayor y menor indice por año es:
# Crear un DataFrame vacío con las columnas necesarias
df_MyM_indice = pd.DataFrame(columns=['anio', 'Mayor_Indice', 'Pais_Mayor_Indice', 'Menor_Indice', 'Pais_Menor_Indice'])

# Recorrer solo años únicos
for anio in df['anio'].unique():
    df_anio = df[df['anio'] == anio]
    df_america = df_anio[df_anio['codigo_iso'].isin(america)]

    # valores que faltan columna indice
    if df_america['indice'].notna().any():  # valores no nulos
        max_indice = df_america.loc[df_america['indice'].idxmax()]
        min_indice = df_america.loc[df_america['indice'].idxmin()]

        # Añadir una nueva fila al DataFrame df_MyM_indice con los valores del año, mayor índice y menor índice

        df_MyM_indice = pd.concat([df_MyM_indice, pd.DataFrame({'anio': [anio], 'Mayor_Indice': [max_indice['indice']],'Pais_Mayor_Indice':[max_indice['pais']], 'Menor_Indice': [min_indice['indice']],'Pais_Menor_Indice':[min_indice['pais']]})], ignore_index=True)
    else:
        #añadir solo el año
        df_MyM_indice = pd.concat([df_MyM_indice, pd.DataFrame({'anio': [anio], 'Mayor_Indice': [None], 'Menor_Indice': [None]})], ignore_index=True)

# Mostrar el DataFrame resultante
df_MyM_indice
```

```python
# Filtrar solo los países de América
df_america = df[df['codigo_iso'].isin(america)]

#mostrar valores maximos y minimos
df_MyM_indice_grouped = df_america.groupby(['anio','pais'])['indice'].agg(['max','min'])

# Mostrar el resultado
df_MyM_indice_grouped
```
